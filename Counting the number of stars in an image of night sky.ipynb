```python
# Required Libraries
import matplotlib.pyplot as plt
from skimage import data
from skimage.feature import blob_dog, blob_log, blob_doh
from math import sqrt 
from skimage.color import rgb2gray
import glob
from skimage.io import imread
```

```python
# Loading the image
sample_image = glob.glob(r"wint_sky.gif") [0]
im = imread(sample_image, as_gray= True)
plt.imshow(im, cmap = 'gray')
plt.show()
```

```python
#Pixel size of image
im.shape
```

```python
# Find the number of stars
# This is the most exciting and critical part of our objective
# The following few commands search for continuous objects in he picture

blobs_log = blob_log(im, max_sigma= 30, num_sigma= 10, threshold= 0.1)
#blobs_log give three outputs for each object found - x-coordinate, y-coordinate, area of the object

blobs_log[:,2] = blobs_log[:,2] * sqrt(2)
num_rows = len(blobs_log)
print("Number of stars counted: ", num_rows)
```

```python
# Validating whether we capture all the stars
# We encircle each estimated star position and check the image if we missed any star

fig, ax = plt.subplots(1,1)
plt.imshow(im, cmap = 'gray')
for blob in blobs_log:
    y,x,r = blob
    c = plt.Circle((x,y), r+4, color = 'yellow', linewidth = 2, fill = False)
    ax.add_patch(c)

```
